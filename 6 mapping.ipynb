```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
super_frame = pd.read_csv('df_resolute.csv', keep_default_na=False)
# convert to datetime object
super_frame['ModifiedOn'] = pd.to_datetime(super_frame['ModifiedOn'])
```

```python
reso_map = folium.Map(location=[37.562120, 24.648375], zoom_start=6)
marker_cluster = MarkerCluster().add_to(reso_map)
```

```python
locations = super_frame[["Latitude","Longitude"]].head(1000)
locationlist = locations.values.tolist()
```

```python
# point cluster
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=str(super_frame['PriceEuro'][point])).add_to(marker_cluster)
```
