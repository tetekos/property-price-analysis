import libraries

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from folium import plugins
```

data

```python
super_frame = pd.read_csv('df_resolute.csv', keep_default_na=False)
# convert to datetime object
super_frame['ModifiedOn'] = pd.to_datetime(super_frame['ModifiedOn'])
```

keep only the ones within GR

```python
super_frame = super_frame[((super_frame.Longitude.between(19.372358, 29.645697)) &\
    (super_frame.Latitude.between(34.80069, 41.748009)))]
super_frame.dropna(subset=['Latitude', 'Longitude'],inplace=True)
super_frame.reset_index(inplace=True)
```

```python
reso_map = folium.Map(location=[super_frame['Latitude'].mean(), super_frame['Longitude'].mean()], zoom_start = 9)
marker_cluster = MarkerCluster().add_to(reso_map)
```

```python
locations = super_frame[["Latitude","Longitude"]].head(2000)
locationlist = locations.values.tolist()
```

```python
# point cluster
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=str(super_frame['PricePerM2Euro'][point])).add_to(marker_cluster)
```

```python
reso_map#.save('spitogatos.html')
```

```python
houses_heatmap = folium.Map(location = [super_frame['Latitude'].mean(), super_frame['Longitude'].mean()], zoom_start = 9)
houses_heatmap.add_child(plugins.HeatMap([[row["Latitude"], row["Longitude"]] for name, row in super_frame.head(10000).iterrows()],
                                        radius=10))

houses_heatmap#.save('spitogatos_heatmap.html')
```
